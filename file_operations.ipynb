Using your knowledge of Python and Matplotlib write your code to transform and graph the data below.

After your code create a new Markdown block to describe your analysis

````python
import matplotlib.pyplot as plt
States =("AK","AL","FL","NY","TX","WY","LA","NJ","OK","NV")
st_NY = 0
st_AL = 0
st_FL = 0
st_TX = 0
st_AK = 0
st_LA = 0
st_WY = 0
st_NJ = 0
st_OK = 0
st_NV = 0
myfile = open("states.txt")

for x in myfile.readlines():
    if x=="NY\n":
        st_NY+=1
    elif x =="AL\n" :
        st_AL+=1  
    elif x =="FL\n"  :
        st_FL+=1
    elif x == "TX\n"   :
        st_TX+=1
    elif x =="NJ\n"     :
        st_NJ+=1

height =[st_NY]

plt.bar(States,height)
plt.xlabel("States")
plt.ylabel("Numbers")
plt.title("How many states is it?")
````
